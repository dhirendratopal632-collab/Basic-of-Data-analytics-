**1**. Define the role of a Data Analyst. What value do they bring to an organization
- A Data Analyst is a person who collects, cleans, and studies data to help organizations make better decisions. They look at patterns, trends, and numbers to explain what is happening and why.
They bring value by

Helping businesses understand their performance

Identifying problems and opportunities

Supporting decision-making with facts instead of guesses

Improving efficiency, sales, and customer satisfaction

**4**. What is Prompt Engineering? Give two types of prompts.

-Prompt Engineering is the skill of writing good instructions for AI tools (like ChatGPT) so they give accurate and useful answers.

1. Zero-Shot Prompt

The AI is asked a question without giving any examples.
Example: “Explain machine learning in simple words.”

2. Few-Shot Prompt

The AI is given a few examples so it understands the pattern.

**5**. Differences between a Business Analyst and a Data Analyst

-Business Analyst Sentences

A Business Analyst mainly focuses on understanding business needs, business processes, and overall business strategy.

The purpose of a Business Analyst is to improve business processes and gather clear business requirements.

A Business Analyst uses skills such as communication, documentation, and requirement gathering.

The outputs of a Business Analyst include reports, business requirement documents, and process maps.

A Business Analyst works closely with stakeholders, project teams, and customers.

Data Analyst – Sentences

A Data Analyst mainly focuses on data, numbers, and trends to understand what is happening in the business.

The purpose of a Data Analyst is to analyze data and provide insights that support decision-making.

A Data Analyst uses skills such as data cleaning, SQL, statistics, and data visualization.

The outputs of a Data Analyst include dashboards, data insights, and trend reports.

A Data Analyst works with databases, analytical tools, and different data sources.

**8**. Explain the evolution of analytics, highlighting the different stages from Descriptive to Generative AI.

-Evolution of Analytics (Descriptive to Generative AI)
1. Descriptive Analytics

Tells what happened in the past using reports and dashboards.

2. Diagnostic Analytics

Tells why it happened using deeper analysis and relationships.

3. Predictive Analytics

Uses machine learning to predict what will happen next.

4. Prescriptive Analytics

Suggests what actions to take for the best results.

5. Generative AI

Uses advanced AI models to create new content, such as text, images, code, or data.

Helps automate decision-making, report writing, and even data preparation.
